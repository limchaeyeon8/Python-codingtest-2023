## 재귀호출 Recursive Call

재귀함수 장점: 코드를 축약할 수 있다

```python
count = 10

def openBox():
    global count
    print('종이상자를 엽니다')
    count -= 1
    if count == 0:
        print('반지를 넣고 반환합니다====')
        return
    openBox()       # 자기자신을 다시 호출
    print('종이상자를 닫습니다')
```

```python
openBox()
```

### 재귀호출하지 않고 반복문으로 덧셈계산

```python
sumValue = 0

for n in range(10, 0, -1):
    sumValue += n

print('10+9+8+7+...+1 = ', sumValue)
```

### 재귀호출로 덧셈

```python
def addNumber(num):
    if num <= 1:
        return 1
    else:
        return num + addNumber(num - 1)
    
print(addNumber(100))
```

### 팩토리얼 반복문

```python
factVal = 1
for n in range(10, 0, -1):
    factVal *= n

print('10*9*8*...*2*1 = ', factVal)
```

### 팩토리얼 재귀호출

```python
def factorial(num):
    if num <= 1:
        return 1
    else:
        return num * factorial(num - 1)
    
print('1558! = ', factorial(1558))
```

### 구구단 재귀호출

```python
def gugudan(dan, num):
    print(f'「 {dan} x {num} = {dan * num:2d}', end=' 」 ')
    if num < 9:
        gugudan(dan, num+1)

for dan in range(2, 10):
    print(f'\n{dan}단 시작 ____')
    gugudan(dan, 1)
```

### n제곱

```python
tab = ' '
def pow(x,n):
    global tab
    tab += '  '
    if n == 0:
        return 1
    print(tab, f'{x}x{x}^{n}-{1}')
    return x * pow(x, n-1)

print('2^10')
print('답 >>>', pow(2, 10))
```

### 피보나치 수 재귀호출

```python
def fibo(n):
    if   n == 0:  return 0
    elif n ==1: return 1
    else:
        return fibo(n-1) + fibo(n-2)
    
print('피보나치수 >>> 0 1', end= ' ')
for i in range(2, 11):          # 10까지 더함
    print(fibo(i), end=' ')
```

```python
def palindrome(pStr):           # kayak
    if len(pStr) <= 1:
        return True
    
    if pStr[0] != pStr[-1]:     # k == k | a == a | y
        return False
    else:
        return palindrome(pStr[1:len(pStr)-1])
    
# ex>    
a = 'kayak'
a[2:len(a)-2]
```

```python
# 전역변수
strArr = ['reaver', 'kayak', 'Borrow or rob', '주유소의 소유주',
          '야 너 이번 주 주번이 너야', '살금 살금']
```

```python
for testStr in strArr:
    print(testStr, end=' -> ')
    testStr = testStr.lower().replace(' ', '')        # 영어대문자 삭제 / 공백 삭제

    if palindrome(testStr) == True:
        print('●회문입니다')
    else:
        print('▲회문 아니오')
```
